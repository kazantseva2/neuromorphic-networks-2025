```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
from scipy.interpolate import interp1d
```

```python
df = pd.read_csv('LiNbO3.csv', sep='\t')  

# Проверяем названия колонок
print("Колонки в файле:", df.columns.tolist())

# Если названия колонок отличаются, переименуем их
df.columns = ['t', 'v', 'c']

# Создание графиков
plt.figure(figsize=(15, 5))

# График V(t)
plt.subplot(1, 2, 1)
plt.plot(df['t'], df['v'], 'b-', linewidth=2, marker='o', markersize=3)
plt.xlabel('Время (с)')
plt.ylabel('Напряжение (V)')
plt.title('V(t)')
plt.grid(True)
plt.ticklabel_format(axis='x', style='sci', scilimits=(0,0))

# График I(V)
plt.subplot(1, 2, 2)
plt.plot(df['v'], df['c'], 'r-', linewidth=2, marker='s', markersize=3)
plt.xlabel('Напряжение (V)')
plt.ylabel('Ток (A)')
plt.title('I(V)')
plt.grid(True)
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('memristor_results.csv', sep=',')  

# Проверяем названия колонок
print("Колонки в файле:", df.columns.tolist())

# Если названия колонок отличаются, переименуем их
df.columns = ['t', 'V', 'I', 'x']

# Создание графиков
plt.figure(figsize=(15, 5))

# График V(t)
plt.subplot(1, 2, 1)
plt.plot(df['t'], df['V'], 'b-', linewidth=2, marker='o', markersize=3)
plt.xlabel('Время (с)')
plt.ylabel('Напряжение (V)')
plt.title('V(t)')
plt.grid(True)
plt.ticklabel_format(axis='x', style='sci', scilimits=(0,0))

# График I(V)
plt.subplot(1, 2, 2)
plt.plot(df['V'], df['I'], 'r-', linewidth=2, marker='s', markersize=3)
plt.xlabel('Напряжение (V)')
plt.ylabel('Ток (A)')
plt.title('I(V)')
plt.grid(True)
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Чтение исходных данных
df_exp = pd.read_csv('Pt-HfO2-TiN.csv', sep='\t')  
df_exp.columns = ['t', 'v', 'c']

# Чтение результатов моделирования
df_sim = pd.read_csv('memristor_results.csv', sep=',')  
df_sim.columns = ['t', 'V', 'I', 'x']

# График ВАХ
plt.figure(figsize=(10, 6))
plt.plot(df_exp['v'], df_exp['c'], 'bo-', linewidth=1, markersize=3, label='Эксперимент')
plt.plot(df_sim['V'], df_sim['I'], 'r-', linewidth=2, label='Моделирование')
plt.xlabel('Напряжение (V)')
plt.ylabel('Ток (A)')
plt.title('ВАХ: эксперимент vs моделирование')
plt.grid(True)
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
plt.legend()
plt.show()
```
